# 1- Import Libraries 

```python
import numpy as np # numpy
import matplotlib.pyplot as plt # matplotlib
import cv2 # opencv

```

# 2- Read the img and turn it into grayscale

```python
img = cv2.imread('1(2).tif') # read image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # BGR ==> GRAY
```

# 3- Function to display image

```python
def show_image(img):
    """
    This Method to disply image in 10*10 scale
    """
    fig = plt.figure(figsize=(10,10))
    ax  = fig.add_subplot(1,1,1, xticks=[], yticks=[])
    ax.imshow(img, cmap='gray')
    
```

# 4- Display the image and image shape

# 5- Sobel edge detection without Blur

```python
sobelX = cv2.Sobel(src=gray, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)  # X edge detection

sobelY = cv2.Sobel(src=gray, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)  # Y edge detection
```

```python
show_image(sobelX)
```

```python
show_image(sobelY)
```

```python
sobel = sobelX +sobelY # combine them togather
show_image(sobel)
```

# 6- Sobel edge detection with Blur

```python
blur = cv2.blur(gray, (5,5))
```

```python
show_image(blur)
```

```python
sobelX = cv2.Sobel(src=blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)  # X edge detection

sobelY = cv2.Sobel(src=blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)  # Y edge detection
```

```python
show_image(sobelX)
```

```python
show_image(sobelY)
```

```python
sobel = sobelX +sobelY # combine them togather
show_image(sobel)
```

# 8- Canny edge detection without Blur

```python
min_threshs = [75,100, 128, 150, 175, 200] #list of min threshs
max_threshs = [200, 210,220,230,240,250]#list of max threshs
```

```python
for i in range(len(min_threshs)):
    Canny = cv2.Canny(gray, min_threshs[i],max_threshs[i])
    show_image(Canny)
```

# 9- Canny edge detection with Blur

```python
for i in range(len(min_threshs)):
    Canny = cv2.Canny(blur, min_threshs[i],max_threshs[i])
    show_image(Canny)
```
